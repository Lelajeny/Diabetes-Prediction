```python
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
```

```python
train_data = pd.read_csv('train.csv')
test_data = pd.read_csv('test.csv')
```

```python
train_data.head()
```

```python
def alphabet_to_number(char):
    if char.isalpha():
        return ord(char.upper()) - ord('A') + 1
    else:
        return char
train_data['CLASS'] = train_data['CLASS'].apply(lambda x: ''.join(map(str, [alphabet_to_number(char) for char in x])))
```

```python
train_data.head()
```

```python
def alphabet_to_number(char):
    if char.isalpha():
        return ord(char.upper()) - ord('A') + 1
    else:
        return char
test_data['CLASS'] = test_data['CLASS'].apply(lambda x: ''.join(map(str, [alphabet_to_number(char) for char in x])))
```

```python
x =train_data.drop(['CLASS'],axis=1)
x
```

```python
y = train_data['CLASS']
y
```

```python
x_test =test_data.drop(['CLASS'],axis=1)
x_test
```

```python
rf_model = RandomForestClassifier(n_estimators=50) 
```

```python
rf_model.fit(x, y)
```

```python
rf_predictions = rf_model.predict(x_test)
```

```python
svm = SVC(kernel='linear', C=2)
```

```python
svm.fit(x, y)
```

```python

svm_predictions = svm.predict(x_test)
```

```python
rf_accuracy = rf_model.score(x_test, y)
svm_accuracy = svm.score(x_test, y)
```

```python
print('Random Forest accuracy:', rf_accuracy)
print('SVM accuracy:', svm_accuracy)
```
